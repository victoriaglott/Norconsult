```python
df["auctioneerID"] = df['auctioneerID'].fillna(df["auctioneerID"].value_counts().index[0])

x = df.drop(columns = ignored_features)
x = x.drop(columns = ["SalePrice"])
y = df["SalePrice"]

```

```python
x.columns
```

```python
#Change format of saledate
```

```python
x["saledate"] = x["saledate"].astype(str)
x_date = x["saledate"].str.split("-",n = 3, expand = True)
x_date
x["SaleYear"] = x_date[0].astype(int)
x["SaleMonth"] = x_date[1].astype(int)
x["SaleDay"] = x_date[2].astype(int)
x = x.drop(columns = ["saledate"])
```

```python
min(x["YearMade"])
```

```python
x.loc[df.YearMade<1950, 'YearMade'] = np.nan
```

```python
x["YearMade"] = x['YearMade'].fillna(df["YearMade"].median())
```

```python
x["YearsOld"] = x["SaleYear"] - x["YearMade"] + x['SaleMonth']/12
```

```python
x['Tire_Size'] = x['Tire_Size'].str.replace(r'\D+','')
x['Tire_Size'] = x['Tire_Size'].replace('', np.nan)
x["Tire_Size"] = x["Tire_Size"].astype(float)
```

```python
x["ProductSize"] = x["ProductSize"].fillna('PS_nan')
```

```python
x['Transmission'] = x['Transmission'].fillna('T_nan')
```

```python
#Split into numerical and categorical features
```

```python
x_numerical = x.select_dtypes(['number'])
```

```python
x_categorical = x.select_dtypes(['object'])
```

```python
1-x_categorical.isnull().sum()/len(x_categorical)
```

```python
#x_numerical = x_numerical.replace({'YearMade': np.nan}, x_numerical['YearMade'].mean() )
```

```python
x_categorical_imputed = x_categorical.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
(1-x_numerical.isnull().sum()/len(x_numerical))
```

```python
(1-x_categorical_imputed.isnull().sum()/len(x_categorical_imputed))
```

```python
x_categorical_imputed = pd.get_dummies(x_categorical_imputed, prefix_sep='_', drop_first=True)
```

```python
x = pd.concat([x_numerical, x_categorical_imputed], axis = 1)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=17)
```

```python

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape = [len(x_train.columns)]))
model.add(tf.keras.layers.Dense(20, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(20, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1))
model.compile(optimizer=tf.keras.optimizers.RMSprop(0.01),
          loss='mse',
          metrics=['mse'])

model.fit(x_train, y_train, epochs = 20)#, batch_size=7)
 
```
